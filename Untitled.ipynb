```python
from sklearn.datasets import load_wine
# Para carregar o dataset
from scipy.stats.mstats import gmean,hmean
# Para calcular a média geométrica e harmônica
from scipy.stats import stats
# Para calcular algumas medidas estatísticas
import matplotlib.pyplot as plt
from cycler import cycler
import seaborn as sns
# Para plotagem de gráfica
import numpy as np
# Para calcular média aritmética dentre outras coisas
import pandas as pd
# Para manipulação dos dados
```

```python
def plot_lines(dados, valor_central, tamanho_ponto=7, legendax=0, legenday=0):
    if valor_central in dados:
        posicao_central = dados.index(valor_central)+1
    else:
        dados.append(valor_central)
        dados.sort()
        posicao_central = dados.index(valor_central)+1

    posicoes = np.arange(1,len(dados)+1,1)
    plt.figure(figsize=(8,6))
    plt.plot(posicoes,dados,'o-', color='red',ms=tamanho_ponto)
    plt.plot(posicao_central, valor_central, 'o-',color='blue',ms=tamanho_ponto)
    if legendax:
        plt.xlabel(legendax)
    if legenday:
        plt.ylabel(legenday)
    plt.show()
```

```python
def plot_line_means(dados, mean, gmean, hmean, tamanho_ponto=10, legendax=0, legenday=0):
    posicoes_centrais=[]
    y=[]
    x = dados

    for m in [mean,gmean,hmean]:
        x = np.append(x, m)

    x.sort()
    for m in [mean,gmean,hmean]:
        if m in x:
            temp_var = np.where(x == m)
            temp_var = temp_var[0] + 1
            #print("\ntemp_var[0]: {}\n".format(temp_var[0]))
            posicoes_centrais = np.append(posicoes_centrais, temp_var)

    print("Dados: {}".format(x))
    print("Media aritmética: {:.5f}, Média Geométrica: {:.5f}, Média Harmonica: {:.5f}".format(mean,gmean,hmean))

    y = np.arange(1,len(x)+1,1)

    plt.figure(figsize=(16,12))
    plt.rc('axes', prop_cycle=(cycler('color', ['r', 'g', 'b'])))
    plt.plot(y,x,'o-', color='black',ms=tamanho_ponto)
    medias = ['MA','MG','MH'] 
    i=0
    for posicao_central,valor_central in zip(posicoes_centrais,[mean,gmean,hmean]):
        plt.plot(posicao_central, valor_central, 'o-',ms=tamanho_ponto)
        plt.text(x=posicao_central+0.2, y=valor_central+0.2, s=medias[i], alpha=0.7, color='#334f8d')
        i+=1
    if legendax:
        plt.xlabel(legendax)
    if legenday:
        plt.ylabel(legenday)
    plt.show()
```

# DATASET WINE
Esse dataset possui 13 colunas e 178 linhas

As colunas são descritas a seguir:
1. Alcohol (Álcool)
2. Malic acid (Ácido málico)
3. Ash (Cinzas/Pó)
4. Alcalinity of ash (Alcalinidade das cinzas)
5. Magnesium (Magnésio)
6. Total phenols (Fenóis totais)
7. Flavanoids (Flavonóides)
8. Nonflavanoid phenols (Fenóis não flavonóides)
9. Proanthocyanins (Proantocianinas)
10. Color intensity (Intensidade da cor)
11. Hue (Matiz)
12. OD280/OD315 of diluted wines (DO280/OD315 de vinhos diluídos)
13. Proline (Prolina)

Essas observações são resultado de uma análise química de vinhos cultivados numa região da itália por 3 produtores diferentes.
